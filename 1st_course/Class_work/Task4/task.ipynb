 Возьмем закон распределения скоростей молекул задаваемой формулой Максвелла

Будем использовать распределение Максвелла f(x) = sqrt(2/pi)* x**2 * exp(-x**2/2)

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as sts
from numpy import var
%matplotlib inline
```

```python
# Произведем выборку из 1000 элементов и посмторим график и гистограмму плотности распределения

maxw = sts.maxwell()

viborka = maxw.rvs(size=1000)

x = np.linspace(0,10,1000)

pdf = maxw.pdf(x)

plt.hist(viborka, normed = True, label='Chart')
plt.plot(x,pdf, label='Density')
plt.ylabel('Number of samples')
plt.xlabel('$x$')
plt.legend()
```

График по выборке

```python
#Среднее из 5
math_o = 2*math.sqrt(2/math.pi)  # теоритическое мат ожидание данного распределения
disp = (24-64/math.pi)           # теоритическая десперсия данного распределения 
n=5
znach = [maxw.rvs(size=5) for i in range(1000)]
middle= [sum(i)/5.0 for i in znach]


mat_o=0
for i in range(1000):
    mat_o = mat_o + middle[i]*(1/1000.)


x = np.linspace(0,5,1000)

norm = sts.norm(loc=math_o, scale=disp/n )
pdf = norm.pdf(x)                                 
plt.hist(middle, normed = True, label='Practical')
plt.plot(x,pdf, label='Theory')
plt.ylabel('Number of samples')
plt.xlabel('$x$')
```

Распределений выборочных средних:
