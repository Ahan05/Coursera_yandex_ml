```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift, estimate_bandwidth
%matplotlib inline
```

```python
data = pd.read_csv('checkins.csv', header=0, sep='\|')
```

```python
data.columns = ['id','user_id','venue_id','latitude','longitude','created_at']
```

```python
data= data.apply(pd.to_numeric, errors='ceries')
```

```python
data = data.fillna('asd')
data = data[data.latitude != 'asd']
data = data[data.id != 'asd']
data = data[data.longitude != 'asd']
data = data[data.user_id != 'asd']
data = data[data.venue_id != 'asd']
```

```python
data = data.drop('id', 1)
data = data.drop('user_id', 1)
data = data.drop('venue_id', 1)
```

```python
data = data.drop('created_at', 1)
```

```python
ms = MeanShift(bandwidth=0.1)
```
